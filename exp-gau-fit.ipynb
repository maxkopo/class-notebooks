# Exponential and Gaussian Curve fitting

Set up inline graphics and import pylab and curve_fit

```python
%matplotlib inline
from pylab import *
from scipy.optimize import curve_fit
```

Create dummy data

```python
xdata = arange(0,5,0.1)

def fexp(x,a,b,c):
    return a*exp(-b*x)+c

def fgau(x,a,b,c):
    return a*exp(-b*x**2)+c

def ddata(x,ns):
    noise = ns*randn(len(x))
    return fexp(x,2.5,1.3,0.5) + noise

ydata = ddata(xdata,0.2)
```

```python
plot(xdata,ydata)
```

Fit data to exp and gau curves

```python
pe,ce = curve_fit(fexp,xdata,ydata)
pg,cg = curve_fit(fgau,xdata,ydata)
yexp = fexp(xdata,*pe)
ygau = fgau(xdata,*pg)
    
fig,ax=subplots(1,2,figsize=(10,4))
ax[0].plot(xdata,ydata,'-sb',ls=':',label='Data')
ax[0].plot(xdata,yexp,'r-',label='Exp fit')
ax[0].legend()
ax[1].plot(xdata,ydata,'-hg',label='Data')
ax[1].plot(xdata,ygau,'r-',label='Gau fit')
ax[1].legend()
```

Print Fit Parameters

```python
print 'Exp fit params         {0:6.3f} {1:6.3f} {2:6.3f}'.format(*pe)
print 'Exp fit std dev        {0:6.3f} {1:6.3f} {2:6.3f}'.format(*sqrt(diag(ce)))
print 'Gau fit parameters     {0:6.3f} {1:6.3f} {2:6.3f}'.format(*pg)
print 'Gau fit std dev        {0:6.3f} {1:6.3f} {2:6.3f}'.format(*sqrt(diag(cg)))
```
