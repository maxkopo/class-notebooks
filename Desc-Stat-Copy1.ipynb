# Simple Descriptive Statistics

```python
%matplotlib inline
from pylab import *
```

Create dummy data

```python
du = rand(1000)
dn = randn(1000)
```

```python
subplot(1,2,1)
plot(du)
subplot(1,2,2)
plot(dn)
```

Function to print max, min, etc.

```python
def dstat(data):
    print "max = %f" % amax(data)
    print "min = %f" % amin(data)
    print "mean = %f" % mean(data)
    print "median = %f" % median(data)
    print "std dev= %f" % std(data)
```

```python
dstat(du)
```

```python
dstat(dn)
```

```python
subplot(1,2,1)
hist(du);
subplot(1,2,2)
hist(dn);
```
